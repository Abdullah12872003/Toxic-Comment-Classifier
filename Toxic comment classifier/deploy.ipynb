```python
# Load the configuration and weights
import joblib

config = joblib.load("/content/gdrive/MyDrive/mp/text_vectorizer_config.joblib")
weights = joblib.load("/content/gdrive/MyDrive/mp/text_vectorizer_weights.joblib")

# Create a new TextVectorization layer with the loaded config
from tensorflow.keras.layers import TextVectorization

vectorizer = TextVectorization.from_config(config)

# Set the loaded weights to the new layer
vectorizer.set_weights(weights)
```
